```python
from top_artists import artists
from config import sp
import pandas as pd
```

```python
artists
```

```python
genre_count = {}
for entry in artists["genres"]:
    for genre in entry:
        if genre not in genre_count.keys():
            genre_count[genre] = 1
        else:
            genre_count[genre] += 1
            
ser = pd.Series(genre_count)
```

```python
ser = ser.sort_values(ascending=False)
ser
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(16, 8), dpi=200)
plot = sns.barplot(x=ser.index[:15], y=ser.values[:15], palette='magma')
plt.xticks(rotation=45)

plot.set_xticklabels([s.title() for s in ser.index[:15]])

plt.tight_layout()
plt.show();
```
